```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, expon
```

```python
sample = np.random.uniform(size=(100000, 50))
print(sample)
```

```python
data = np.sum(a=sample, axis=1)
print(data)
```

```python
plt.hist(data, 
         bins=100, 
        #  bins='auto', 
         density=True)
plt.show()
```

```python
sample_mean = np.mean(data)
sample_variance = np.var(data, ddof=1)
sample_standart_deviation = np.sqrt(sample_variance)
print(sample_mean, sample_variance, sample_standart_deviation)
```

```python
# x = np.linspace(sample_mean - 3 * sample_standart_deviation, sample_mean + 3 * sample_standart_deviation)
# pdf = norm.pdf(x=x, loc=sample_mean, scale=sample_standart_deviation)
x = np.linspace(sample_mean - 3 * sample_standart_deviation, sample_mean + 3 * sample_standart_deviation)
pdf = norm.pdf(x=x, loc=sample_mean, scale=sample_standart_deviation)
```

```python
plt.hist(data, 
         bins=30, 
        #  bins='auto', 
         density=True)
plt.plot(x, pdf)
plt.show()
```
