```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Define distribution parameters
mu = 106  # population mean
sigma = 17  # population standard deviation
```

```python
# Define the distribution
dist = stats.norm(loc=mu, scale=sigma)
```

```python
# Choose a sample size
n = 420
```

```python
a = dist.rvs(n)
```

```python
# Level of statistical significance
alpha = 0.05
```

```python
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
y = dist.pdf(x)
```

```python
# two-sided
left_bound = dist.ppf(alpha / 2)
right_bound = dist.ppf(1 - alpha / 2)
print(left_bound, right_bound)
# print(dist.cdf(right_bound) - dist.cdf(left_bound))
print(f"P(T <= l) = {dist.cdf(left_bound)}, P(T >= r) = {1 - dist.cdf(right_bound)}")
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("Value")
ax.set_ylabel("Probability density")
ax.set_title(fr"$ N(\mu={mu:.0f}$, $\sigma={sigma:.0f}) $")
ax.grid()
ax.fill_between(x=x,
                y1=y, 
                where=(x <= left_bound) | (right_bound <= x),
                color="r",
                alpha=0.5)
plt.show()
```

```python
# right-sided
right_bound = dist.ppf(1 - alpha)
print(right_bound, 1 - dist.cdf(right_bound))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("Value")
ax.set_ylabel("Probability density")
ax.set_title(fr"$ N(\mu={mu:.0f}$, $\sigma={sigma:.0f}) $")
ax.grid()
ax.fill_between(x=x,
                y1=y, 
                where=(right_bound <= x),
                color="r",
                alpha=0.5)
plt.show()
```

```python
# left-sided
left_bound = dist.ppf(alpha)
print(left_bound, dist.cdf(left_bound))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("Value")
ax.set_ylabel("Probability density")
ax.set_title(fr"$ N(\mu={mu:.0f}$, $\sigma={sigma:.0f}) $")
ax.grid()
ax.fill_between(x=x,
                y1=y, 
                where=(x <= left_bound),
                color="r",
                alpha=0.5)
plt.show()
```
