```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
normal = np.random.standard_normal(size=10000)
```

```python
normal_means = np.zeros_like(normal)
for i in range(10000):
    normal_means[i] = np.mean(normal[:i + 1])
```

```python
plt.plot(normal_means, label="Normal")
plt.grid()
plt.show()
```

```python
cauchy = np.random.standard_cauchy(size=100000)
```

```python
cauchy_means = np.zeros_like(cauchy)
for i in range(100000):
    cauchy_means[i] = np.mean(cauchy[:i + 1])
```

```python
plt.plot(cauchy_means, label="Normal")
plt.grid()
plt.show()
```
