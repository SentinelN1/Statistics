```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sample_size = 10000
n = 100
m = 0
s = 1
sample = np.random.normal(loc=m, scale=s, size=(sample_size, n))
```

```python
mean_estimator = np.mean(sample, axis=1)
print(np.mean(mean_estimator))
```

```python
np.mean(mean_estimator), np.var(mean_estimator)
```

```python
mean_dist = stats.norm(loc=m, scale=s / np.sqrt(n))
x = np.linspace(m - 3 * s / np.sqrt(n), m + 3 * s / np.sqrt(n))
y = mean_dist.pdf(x)
```

```python
plt.title('Mean estimator')
plt.hist(mean_estimator, bins='auto', density=True, histtype='stepfilled', label='$ \hat{\mu} $')
plt.plot(x, y, label='$ N(\mu, \\frac{1}{n}\sigma^2) $')
plt.legend()
plt.show()
```

```python
var_estimator = np.var(sample, ddof=0, axis=1)
print(np.mean(var_estimator))
```

```python
dist = stats.chi(df=n - 1)
x = np.linspace(s, 20)
y = dist.pdf(x)
```
