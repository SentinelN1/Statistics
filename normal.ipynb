```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
sample_size = 100
sample = np.round(np.random.normal(loc=182, scale=23, size=sample_size), 2)
```

```python
plt.hist(sample, bins='auto', density=True)
# plt.hist(sample, density=True)
plt.show()
```

```python
sample_mean = np.mean(sample)
# sample_variance = np.var(sample)
sample_variance = np.sum(np.power(sample - sample_mean, 2)) / (sample_size - 1)
sample_standart_deviation = np.sqrt(sample_variance)
print(sample_mean, sample_variance)
```

```python
x = np.linspace(sample_mean - 3 * sample_standart_deviation, sample_mean + 3 * sample_standart_deviation, 1000)
pdf = scipy.stats.norm.pdf(x, loc=sample_mean, scale=sample_standart_deviation)
plt.hist(sample, bins='auto', density=True)
plt.plot(x, pdf)
plt.show()
```

```python
t_score = (sample_mean - 182) / np.sqrt(sample_variance / sample_size)
print(t_score)
```

```python
res = scipy.stats.ecdf(sample)
ecdf = res.cdf.evaluate(x)
cdf = scipy.stats.norm.cdf(x, loc=sample_mean, scale=sample_standart_deviation)
plt.plot(x, ecdf, linestyle='-', label = 'Empirical')
plt.plot(x, cdf, label = 'Theoretical')
plt.show()
```

```python
mean_abs_error = np.mean(np.abs(ecdf - cdf))
mean_sqare_error = np.sqrt(np.mean(np.power(ecdf - cdf, 2)))
print(mean_abs_error, mean_sqare_error)
```

```python
sample_size = 10000
n = 100
sample = np.sum(
    np.random.uniform(
        low=0,
        high=1,
        size=(
            n,
            sample_size
        )
    ), 
    axis=1)
```

```python
plt.hist(sample, bins='auto', density=True)
plt.show()
```

```python
sample_mean = np.mean(sample)
sample_variance = np.sum(np.power(sample - sample_mean, 2)) / (sample_size - 1)
# sample_variance = np.var(sample)
sample_standart_deviation = np.sqrt(sample_variance)
print(sample_mean, sample_variance)
```

```python
x = np.linspace(sample_mean - 3 * sample_standart_deviation, sample_mean + 3 * sample_standart_deviation)
```

```python
res = scipy.stats.ecdf(sample)
ecdf = res.cdf.evaluate(x)
cdf = scipy.stats.binom.cdf(x, n, 0.5)
plt.plot(x, ecdf)
plt.plot(x, cdf)
plt.show()
```

```python
mean_abs_error = np.mean(np.abs(ecdf - cdf))
mean_sqare_error = np.sqrt(np.mean(np.power(ecdf - cdf, 2)))
print(mean_abs_error, mean_sqare_error)
```
