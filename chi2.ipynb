```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
num_trials = 10000
n = 50
```

```python
sample = np.sum(np.power(np.random.normal(size=(num_trials, n)), 2), axis=1)
```

```python
plt.hist(sample, bins='auto', density=True)
plt.xlabel('Random value')
plt.ylabel('Probability density')
plt.show()
```

```python
dist = stats.chi2(n)
x = np.linspace(sample.min(), sample.max())
y = dist.pdf(x)
```

```python
plt.hist(sample, bins='auto', histtype='stepfilled', density=True)
plt.plot(x, y, label='$ \chi^{2}_{n} $')
plt.xlabel('Random value')
plt.ylabel('Probability density')
plt.legend()
plt.show()
```
