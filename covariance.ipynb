```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize, stats, special
```

```python
n = 50
noise = np.random.normal(size=n)
x = np.linspace(-1, 1, num=n)
# x = np.random.uniform(size=n)
y = x**5 - x**2 + 3 * x + 2 + noise
```

```python
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
np.cov(x, y), np.corrcoef(x, y)
```
