```python
data = np.log(sample)
print(data)
```

```python
plt.hist(data, bins=50, density=True)
plt.show()
```

```python
sample_mean = np.mean(data)
sample_variance = np.var(data)
sample_standart_deviation = np.sqrt(sample_variance)
print(sample_mean, sample_variance, sample_standart_deviation)
```
